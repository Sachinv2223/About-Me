```python
import pandas as pd
```

```python
# Load the matches data
matches_data = {
    'home_team': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'away_team': ['B', 'C', 'D', 'A', 'F', 'G', 'H', 'E'],
    'home_score': [3, 1, 2, 0, 2, 1, 3, 0],
    'away_score': [1, 2, 1, 0, 1, 1, 0, 2]
}
matches = pd.DataFrame(matches_data)
matches
```

```python
# Load the guest teams data
guest_teams_data = {
    'team_name': ['B', 'D', 'F', 'H']
}
guest_teams = pd.DataFrame(guest_teams_data)
guest_teams
```

```python
# Function to determine if a team is a guest team
def is_guest_team(team_name):
    return team_name in guest_teams['team_name'].values
```

```python
# Calculate points based on match results
def calculate_points(home_score, away_score, home_team, away_team):
    if home_score > away_score:
        home_points = 3
        away_points = 0
    elif home_score < away_score:
        home_points = 0
        away_points = 3
    else:
        home_points = 1
        away_points = 1

    if is_guest_team(home_team):
        home_points = home_points + 1 if home_points > 0 else 0
    if is_guest_team(away_team):
        away_points = away_points + 1 if away_points > 0 else 0

    return home_points, away_points
```

```python
# Apply the points calculation to the matches DataFrame
matches[['home_points', 'away_points']] = matches.apply(lambda row: calculate_points(row['home_score'], row['away_score'], row['home_team'], row['away_team']), axis=1, result_type='expand')
matches
```

```python
# Aggregate points for each team
home_team_points = matches.groupby('home_team')['home_points'].sum().reset_index()
away_team_points = matches.groupby('away_team')['away_points'].sum().reset_index()
total_points = pd.merge(home_team_points, away_team_points, left_on='home_team', right_on='away_team', how='outer').fillna(0)
total_points['total_points'] = total_points['home_points'] + total_points['away_points']
total_points = total_points[['home_team', 'total_points']].rename(columns={'home_team': 'team_name'})
```

```python
# Sort teams by points
ranked_teams = total_points.sort_values(by='total_points', ascending=False).reset_index(drop=True)
```

```python
# Print the ranked teams
print("Ranked Teams:")
ranked_teams
```
